## Split the Data into Training and Testing Sets

```python
# Create the labels set `y` and features DataFrame `X`
X = data.drop(columns='spam')
y = data['spam']
```

```python
# Check the balance of the labels variable (`y`) by using the `value_counts` function.
y.value_counts()
```

```python
# Split the data into X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train.head()
```

## Scale the Features

Use the `StandardScaler` to scale the features data. Remember that only `X_train` and `X_test` DataFrames should be scaled.

```python
from sklearn.preprocessing import StandardScaler

# Create the StandardScaler instance
scalar = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
scalar.fit(X_train)
```

```python
# Scale the training data
X_train_scaled = scalar.transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scalar.transform(X_test)
X_test_scaled
```

## Create and Fit a Logistic Regression Model

Create a Logistic Regression model, fit it to the training data, make predictions with the testing data, and print the model's accuracy score. You may choose any starting settings you like. 

```python
# Train a Logistic Regression model and print the model score
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=1)
lr.fit(X_train_scaled, y_train)
```

```python
# Make and save testing predictions with the saved logistic regression model using the test data
y_pred_lr = lr.predict(X_test_scaled)
# Review the predictions
y_pred_lr
```

```python
# Calculate the accuracy score by evaluating `y_test` vs. `testing_predictions`.
accuracy_score(y_test, y_pred_lr)
```

## Create and Fit a Random Forest Classifier Model

Create a Random Forest Classifier model, fit it to the training data, make predictions with the testing data, and print the model's accuracy score. You may choose any starting settings you like. 

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=150, random_state=1).fit(X_train_scaled, y_train)
rf
```

```python
# Make and save testing predictions with the saved logistic regression model using the test data
y_pred_rf = rf.predict(X_test_scaled)

# Review the predictions
y_pred_rf
```

```python
# Calculate the accuracy score by evaluating `y_test` vs. `testing_predictions`.
accuracy_score(y_test, y_pred_rf)
```

## Additional evaluation metrics

Calculate following metrics:
- Precision
- Recall
- F1 Score
- ROC AOC

```python
def evaluate_model(y_test, y_pred):
    precision = precision_score(y_test, y_pred, pos_label=1)  #pos_label=1 indicates msg has been flagged as spam
    recall = recall_score(y_test, y_pred, pos_label=1)
    f1 = f1_score(y_test, y_pred, pos_label=1)
    return precision, recall, f1
```

```python
lr_metrics = evaluate_model(y_test, y_pred_lr)
print("Logistic Regression: Precision={}, Recall={}, F1={}".format(*lr_metrics))

rf_metrics = evaluate_model(y_test, y_pred_rf)
print("Random Forest Classifier: Precision={}, Recall={}, F1={}".format(*rf_metrics))
```

```python
# Create a classification report for Logistic Regression model
print(classification_report(y_test, y_pred_lr, labels = [1, 0]))
```

```python
# Create a classification report for Random Forest classification model
print(classification_report(y_test, y_pred_rf, labels = [1, 0]))
```

```python
# Calculate the balanced accuracy score
print(f'Linear Regression balanced accuracy = {balanced_accuracy_score(y_test, y_pred_lr)}')
print(f'Random Forest Classification balanced accuracy = {balanced_accuracy_score(y_test, y_pred_rf)}')
```

```python
# Confusion Matrix for Linear Regression Model
confusion_matrix(y_test, y_pred_lr, labels = [1,0])
```

```python
# Confusion Matrix for Random Forest Classifier Model
confusion_matrix(y_test, y_pred_rf, labels = [1,0])
```

```python
# Calculate ROC AUC score for Logistic Regression Model
# Predict values with probabilities
pred_probas = lr.predict_proba(X_test)

# We only need the predictions for the 1 class; use a list comprehension to 
# gather the second value from each list
pred_probas_firsts = [prob[1] for prob in pred_probas]

# Calculate the roc_auc_score for Linear Regression Model
print(f'Linear Regression ROC AUC score = {roc_auc_score(y_test, pred_probas_firsts)}')

#lr_roc_auc = roc_auc_score(y_test, lr.predict_proba(X_test)[:, 1])
```

```python
# Calculate ROC AUC score for Random Forest Classifier Model
# Predict values with probabilities
pred_probas = rf.predict_proba(X_test)

# We only need the predictions for the 1 class; use a list comprehension to 
# gather the second value from each list
pred_probas_firsts = [prob[1] for prob in pred_probas]

# Calculate the roc_auc_score for Random Fortest Model
print(f'Random Forest Classifier ROC AUC score = {roc_auc_score(y_test, pred_probas_firsts)}')

#rf_roc_auc = roc_auc_score(y_test, rf.predict_proba(X_test)[:, 1])
```
